# Random Forest for selecting important features

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
att_train.isna().sum()
```

There are null values in Num of Companies Worked and total working years. SO, as they are continous we will replace it by mean.

```python
new_var = np.where(att_train['NumCompaniesWorked'].isnull(),att_train['NumCompaniesWorked'].mean(),att_train['NumCompaniesWorked'])
att_train['NumCompaniesWorked']=new_var
new_var2 = np.where(att_train['TotalWorkingYears'].isnull(),att_train['TotalWorkingYears'].mean(),att_train['TotalWorkingYears'])
att_train['TotalWorkingYears']=new_var
att_train.isnull().sum().sum()
```

```python
objList = att_train.select_dtypes(include = "object").columns
print (objList)
```

```python
#Label Encoding for object to numeric conversion
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in objList:
    att_train[feat] = le.fit_transform(att_train[feat].astype(str))

att_train.head(10)
```

```python
rf_model=RandomForestClassifier(n_estimators=1000,max_features=2,oob_score=True)
```

```python
features=['Age','BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EducationField', 'EmployeeCount', 'EmployeeID', 'Gender',
       'JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'Over18', 'PercentSalaryHike', 'StandardHours',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
```

```python
rf_model.fit(X=att_train[features],y=att_train['Attrition'])
```

```python
print(rf_model.oob_score_)
```

So, the model is 99.95% accurate.

```python
for feature,imp in zip(features,rf_model.feature_importances_):
    print(feature,imp)
```

In this Dataset: Age,DistanceFromHome,MonthlyIncome,PercentSalaryHike,YearsAtCompany Can be considered as imp. features.

```python
imp_features=['Age', 'DistanceFromHome', 'MonthlyIncome','PercentSalaryHike','YearsAtCompany']
rf_model.fit(X=att_train[imp_features],y=att_train['Attrition'])
print(rf_model.oob_score_)
```

Thus, now the model is 100% accurate.

# Classification

```python
tree_model = tree.DecisionTreeClassifier(max_depth=8)
features=['Age', 'DistanceFromHome', 'MonthlyIncome','PercentSalaryHike','YearsAtCompany']
tree_model.fit(X=att_train[features],y=att_train['Attrition'])
```

```python
with open('att_Dtree.dot','w') as f:
    f=tree.export_graphviz(tree_model,feature_names=features,out_file=f)
    
```

Inferences from Classification:

```python
r=tree.export_text(tree_model,feature_names=features)
print(r)
```
