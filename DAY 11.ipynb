# Correlation with Attrition

### Step-1 : Launching and importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset=pd.read_csv("general_data.csv")
dataset.head()
```

### Step-2 Checking for nulls and duplicates

```python
dataset.isnull()
```

```python
dataset.duplicated()
```

```python
dataset.drop_duplicates()
```

### Stage -3 Convert category to numeric for Attrition: Yes-1 and NO-0

```python
dataset["Attrition"] = dataset["Attrition"].astype('category')
dataset.dtypes
```

```python
dataset["Attrition"] = dataset["Attrition"].cat.codes
dataset.head()
```

### 1)Age

```python
from scipy.stats import pearsonr
stats,p = pearsonr(dataset["Attrition"],dataset["Age"])
print(stats,p)
```

Age is negatively correlated to attrition rate. So, the younger employees have higher attrition rate.

## 2)DistanceFromHome

```python
stats,p = pearsonr(dataset["Attrition"],dataset["DistanceFromHome"])
print(stats,p)
```

DistanceFromHome is very weakly negatively correlated to attrition rate. Also, as p-value is > 0.05, we accept H0, i.e DistanceFromHome and attrition rate are not much correlated.
