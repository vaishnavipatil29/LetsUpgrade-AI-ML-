# Naive Bayes Theorem

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import*
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
df = pd.read_csv("train.csv")
df.head()
```

### Convert Categorical into Numerical

```python
objList=['Sex','Embarked']
print (objList)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in objList:
    df[feat] = le.fit_transform(df[feat])

df.head()
```

### D.V and IDV

```python
df=df.drop(['Name','Ticket','Cabin'],axis=1)
DV = ['Survived','Sex','Pclass','SibSp','Parch','Embarked']
#x will be df.drop([y])
df.head()
```

### Spliting Train and test datasets, Applying Naive Bayes Theorem

```python
clf = BernoulliNB()
```

```python
def diag(matrix,s):
    tot =0
    for i in range(0,s):
        tot=tot+ matrix[i][i]
    return tot




for a in DV:
    y=df[a]
    X=df.drop([a,'PassengerId'],axis=1)
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
    y_pred=clf.fit(X_train,y_train).predict(X_test)
    score = accuracy_score(y_test,y_pred,normalize=True)
    matrix= confusion_matrix(y_test,y_pred)
    s=matrix.shape[0]
    correct_records=diag(matrix,s)
    print(a,"-"," Accuracy Score: ",score)
    print("    ","Confusion Matrix: ",matrix)
    print("    ","No. of recorders which are classified correctly are: ",correct_records)
    
    print('\n')
```

Thus, the most accurate records are classified for " Survived " as the Target variable. So, this data is most suitable for Classifying whether a person Survived or Not.
