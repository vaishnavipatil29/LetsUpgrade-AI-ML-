# Project 1: Build Decision Tree(DV-"Survived",IDV-"Age,Gender and Fare") and Prediction

```python
# Importing packages
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn import preprocessing
```

#### CLASSIFICATION

```python
titanic_train = pd.read_csv("train.csv")
titanic_train.head(10)
```

######  Step -1 : Handling Missing Values

```python
titanic_train.isna().sum()
```

Age, Sex and Fare have no null values.

###### Step -2 Convert Sex into numerical

```python
label_encoder=preprocessing.LabelEncoder()
encoded_sex=label_encoder.fit_transform(titanic_train['Sex'])
titanic_train['Sex']=encoded_sex
titanic_train.head(5)
```

###### Step -3 Declaration

```python
tree_model = tree.DecisionTreeClassifier(max_depth=8)
```

###### Step-4 D.V and I.D.V

```python
features = ["Sex","Age","Fare"]
tree_model.fit(X=titanic_train[features],y=titanic_train['Survived'])
```

###### Step-5 Opening a file for 'webgraphviz.com'

```python
with open('Dtree.dot','w') as f:
    f=tree.export_graphviz(tree_model,feature_names=["Sex","Age","Fare"],out_file=f)
    
```

Inferences from Classification:

```python
r=tree.export_text(tree_model,feature_names=features)
print(r)
```

Some of the inferences are-

1) Females with fare>48.2, age<8 : Not Survived.

2) Females with fare : (8.04,10.48] and age <=19 : Survived.

3) Males with Age <=6.50 and Fare <=20.83 : Survived.

4) Males with Age <=3.50  and fare : (20.83,23.54]: Not Survived. etc..

#### Prediction

```python
titanic_test = pd.read_csv("test.csv")
titanic_test.head(5)
```

```python
titanic_test.isna().sum()
```

No Null values are present.

```python
label_encoder=preprocessing.LabelEncoder()
encoded_sex=label_encoder.fit_transform(titanic_test['Sex'])
titanic_test['Sex']=encoded_sex
titanic_test.head(5)
```

```python
features = ["Sex","Age","Fare"]
test_pred = tree_model.predict(X=titanic_test[features])
predicted_output = pd.DataFrame({"PassengerId":titanic_test['PassengerId'],"Survived":test_pred})
predicted_output.to_csv("Output.csv",index=False)
```

```python
b=pd.read_csv("Output.csv")
```

```python
merged = titanic_test.merge(b, on='PassengerId')
merged.to_csv("Prediction.csv", index=False)
```
