# Logistic Regression

DV - Atrrition(Y/N)

```python
import numpy as np
import pandas as pd
df=pd.read_csv("general_data.csv")
df.head(5)
```

```python
df.isna().sum()
```

```python
df['NumCompaniesWorked']=df['NumCompaniesWorked'].fillna(df['NumCompaniesWorked'].mean())
df.isnull().sum().sum()
```

```python
df['TotalWorkingYears']=df['TotalWorkingYears'].fillna(df['TotalWorkingYears'].mean())
df.isnull().sum().sum()
```

```python
objList = df.select_dtypes(include = "object").columns
print (objList)
```

```python
#Label Encoding for object to numeric conversion
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for feat in objList:
    df[feat] = le.fit_transform(df[feat].astype(str))

df.head(10)
```

```python
df.drop_duplicates
```

```python
df.columns
```

```python
df['NumCompaniesWorked'].isna().sum()
```

```python
y=df['Attrition']
X=df[['Age', 'BusinessTravel', 'Department', 'DistanceFromHome',
       'Education', 'EducationField', 'Gender',
       'JobLevel', 'JobRole', 'MaritalStatus', 'MonthlyIncome',
       'NumCompaniesWorked', 'PercentSalaryHike', 
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager']]
```

```python
import statsmodels.api as sm
X1=sm.add_constant(X)
Logistic=sm.Logit(y,X1)
result=Logistic.fit()
result.summary()
```

The features that are significantly important for deciding the attrition(P<0.05):

Age, Department, EducationField	, JobRole , MaritalStatus, NumCompaniesWorked , YearsSinceLastPromotion, YearsWithCurrManager, TotalWorkingYears, TrainingTimesLastYear
